## GRADIENT DESCENT EXAMPLE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# ...we generate a set of points that follow a normal distribution...


mu, sigma, n_sample, n_events = 10, 1, 100, 10000
np.random.seed(0)

gauss_sample = np.random.normal(loc=mu, scale=sigma, size=n_events)
freq, interval_bounds = np.histogram(gauss_sample, bins=n_sample)
 
x = np.empty(shape=n_sample)
iter = np.nditer(interval_bounds, flags=['f_index'])
while not iter.finished:
    if iter.index+1 < interval_bounds.size:
        x[iter.index] = (interval_bounds[iter.index] + interval_bounds[iter.index+1])/2  
    iter.iternext()
    
#print(x)   
#print(freq)

points = pd.DataFrame([x,freq], index = (['x', 'freq'])).T
```

```python

# ...we plot the points to see the sample...


points.plot('x','freq')
#print(points)
```
